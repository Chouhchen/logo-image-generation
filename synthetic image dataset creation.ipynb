```python
## https://medium.com/@alexppppp/how-to-create-synthetic-dataset-for-computer-vision-object-detection-fd8ab2fa5249

from PIL import Image, ImageDraw, ImageFilter

im2 = Image.open('defect1.jpg')
iml = Image.open('logo1.jpg')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
imgplot = plt.imshow(iml)
plt.show()
```

```python
back_im = iml.copy()
mask_im = im2.copy()
back_im.paste(im2,(200, 200), mask_im)
```

```python
imgplot = plt.imshow(back_im)
plt.show()
```

```python
back_im.save(r'\datasets\test1.png')
```

```python
#https://medium.com/@alexppppp/how-to-create-synthetic-dataset-for-computer-vision-object-detection-fd8ab2fa5249
    
obj_dict = {'defect1.jpg':1, 'defect2.jpg':2, 'defect3.jpg':3}
```

```python
for k in obj_dict:
    print(k)
```

```python
for k in obj_dict:
    imD = Image.open(k)
    mask_im = imD.copy()
    back_im.paste(imD,(200, 200), mask_im)
    back_im.save('test_{k}.png')
```
